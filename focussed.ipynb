```python
from ultralytics import YOLO
import os
```

```python
DATASET_PATH = "/content/drive/MyDrive/dataset2"
DATA_YAML = os.path.join(DATASET_PATH, "data.yaml")
```

```python
# Use STRONGER Model
model = YOLO("yolov8m.pt")
```

```python
# Train for High Accuracy
model.train(
    data=DATA_YAML,
    epochs=80,
    imgsz=960,
    batch=8,              # Reduce batch for stability
    optimizer="AdamW",    # optimizer
    lr0=0.001,
    patience=30,
    cos_lr=True,
    degrees=5,
    translate=0.1,
    scale=0.5,
    fliplr=0.5,
    mosaic=1.0,
    mixup=0.2,
    copy_paste=0.1,
    # device="cpu"
)
```
